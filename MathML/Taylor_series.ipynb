### Maclaurin Series

For approximation of function we should know all about a point into 0.

Zero order - constant which equal a function value in 0:
$$
y = f(0) = c
$$
First order - it is linear approximation:
$$
y = f'(0)*x + f(0)
$$
Second order - from this step we need to know more about parabolic function, for instance - first and second derivative:
$$
y = a*x^2 + b*x + c
$$
$$
y' = 2a*x + b
$$
$$
y'' = 2a = f''(0) => a = \frac{f''(0)}{2}
$$
$$
y = f(0) + f'(0)*x + \frac{f''(0)}{2}
$$
For the third order it will be third derivative devided by 6. Now we can rewrite our approximation equation into general form:
$$
g(x) = \sum_{n=0}^{\inf} {\frac{f^{(n)}(0)}{n!} * x^n}
$$
    This formula - Maclaurin Series which gives us posibility to approximate any continuous function

If we talk about Taylor - it is more common form of Maclaurin Series => we can approximate function in any point of space. For doing it we need to consider formulas above again.

Zero order
$$
y = f(p) = c
$$
First order
$$
y = f'(p)*x + c
$$
$$
f(p) = f'(p)*p + c => c = f(p) - f'(p)*p
$$
$$
y = f'(p)*x + f(p) - f'(p)*p => f'(p)*(x - p) + f(p)
$$
Second order will be the same like in the first view but with substitude $x$ to $(x - p)$
$$
y = \frac{f''(p)}{2}*(x - p)^2 + f'(p)*(x - p) + f(p) 
$$
Now we are ready to rewrite Maclaurin formula to Taylor Series:
$$
y = \sum_{n=0}^{\inf}{\frac{f^{(n)}(p)}{n!}*(x - p)^{(n)}}
$$

### Multivariate Taylor

$$
f(x+\Delta x, y+\Delta y) = f(x,y) + (\delta_x f(x,y)\Delta x + \delta_y f(x,y)\Delta y) + \frac{1}{2}(\delta_{xx}f(x,y)\Delta x^2 + 2\delta_{xy}f(x,y)\Delta x \Delta y + \delta_{yy}f(x,y)\Delta y^2)
$$
Second term looks familiar:
$$
[\delta_x f(x,y),\delta_y f(x,y)]*[\Delta x, \Delta y] => J_f * \Delta \vec{x}
$$
Third term is Hessian.
$$
\frac{1}{2} *
\begin{bmatrix}
\Delta x \\
\Delta y
\end{bmatrix} *
\begin{bmatrix}
\delta_xx f(x,y), \delta_xy f(x,y) \\
\delta_yx f(x,y), \delta_yy f(x,y)
\end{bmatrix} *
\begin{bmatrix}
\Delta x \\
\Delta y
\end{bmatrix} => \frac{1}{2} * \Delta \vec{x}^T * H_f * \Delta \vec{x}
