$$

For example:

f(x) = 5x^2 =>

### 1 operation: power rule

$$
    f(x) = x^b => f'(x) = bx^{(b-1)}
$$

Additional mark: if $f(x) = e^x => f'(x) = \frac{d(e^x)}{dx} = e^x$ (*it is from sin/cos rules*)

### 2 operation: summation rule
Derivative of function sum is sum of function derivatives. For example:

$$
    f'(x) = \frac{d(x^3 + e^x)}{dx} = \frac{d(x^3)}{dx} + \frac{d(e^x)}{dx} = 3x^2 + e^x
$$

### 3 operation: product rule
If imagine product of two functions like rectangle where different sides are represented by our functions => product these functions is square of this rectangle. Hence increasing of functions - derivatives. Label our functions f(x) and g(x), product of these functions is p(x).

$$
    \Delta(p(x)) = f(x)*\Delta(g(x)) + \Delta(f(x))*g(x)
$$
For example:
$$
f(x) = x^3
$$
$$
g(x) = e^x => p(x) = x^3 * e^x
$$
$$
p'(x) = \frac{d(x^3 * e^x)}{dx} = \frac{x^3 * d(e^x)}{dx} + \frac{d(x^3)}{dx} * e^x = x^3 * e^x + 3x^2 * e^x
$$

### 4 operation: chain rule
We can simplify any function by spliting big function to small pieces.

If we have this situation: 
$$
h(p(m)) => \frac{d(h(p(m)))}{dm} = \frac{d(h(p))}{dp} * \frac{d(p(m))}{dm}
$$

It is useful with complicated functions, for example:

$$
h(p) = sin(p)
$$
$$
p(m) = m^3 * e^m
$$
$$
\frac{d(h(p(m)))}{dm} = \frac{d(sin(p))}{dp} * \frac{d(m^3 * e^m)}{dm} = cos(p) * (m^3 * e^m + 3m^2 * e^m)
$$
Substitution of p(m) to original function after calculating derivative gives us:
$$
\frac{d(h(p(m)))}{dm} = cos(m^3 * e^m) * (m^3 * e^m + 3m^2 * e^m)
$$

### Example with all 4 operations
$$
f(x) = \frac{sin(2x^5+3x)}{e^{7x}}
$$
1 step - rewrite it:
$$
f(x) = sin(2x^5+3x)*e^{-7x}
$$
2 step - use product rule:
$$
f'(x) = \frac{d(sin(2x^5+3x))}{dx}*e^{-7x}+sin(2x^5+3x)*\frac{d(e^{-7x})}{dx} = f_1(x)+f_2(x)
$$
3 step - use chain rule and summation rule for first part of sum:
$$
p(x) = 2x^5+3x
$$
$$
p'(x) = \frac{d(2x^5)}{dx}+\frac{d(3x)}{dx} = 10x^4+3
$$
$$
t(p) = sin(p)
$$
$$
t'(p) = cos(p) -> cos(2x^5+3x)
$$
    Substitude it to first part of sum using chain rule:
$$
f_1(x) = \frac{d(sin(2x^5+3x))}{dx}*e^{-7x} = \frac{cos(2x^5+3x)*(10x^4+3)}{e^{7x}}
$$
4 step - use power rule and chain rule for second part of sum:
$$
q(x) = -7x
$$
$$
q'(x) = -7
$$
$$
r(q) = e^q
$$
$$
r'(q) = e^q
$$
$$
f_2(x) = sin(2x^5+3x)*\frac{d(e^{-7x})}{dx} = sin(2x^5+3x)*q'(x)*r'(q) = sin(2x^5+3x)*(-7)*e^q = sin(2x^5+3x)*(-7)*e^{-7x}
$$
Full derivative of function:
$$
f'(x) = -7*\frac{sin(2x^5+3x)}{e^{7x}}+\frac{cos(2x^5+3x)*(10x^4+3)}{e^{7x}} = \frac{cos(2x^5+3x)*(10x^4+3)}{e^{7x}}-7*f(x)
$$





## Differentiate with respect to everything
It is often situation when function has few variables, for example weight, speed, etc. Depends on a context of solving problem each of these variables can be subject of interest in some research task, that's why it is useful to be able to calculate partial derivative with respect to every variable in your system. Additional remark here - each variable can be a function of other variable. Let's consider this situation!
$$
f(x,y,z) = sin(x)e^{yz^2}
$$
Let's calculate partial derivatives with respect to every variable:
$$
\frac{\delta{f(x,y,z)}}{\delta{x}} = cos(x)*e^{yz^2}
$$
$$
\frac{\delta{f(x,y,z)}}{\delta{y}} = sin(x)*e^{yz^2}*z^2
$$
$$
\frac{\delta{f(x,y,z)}}{\delta{z}} = sin(x)*e^{yz^2}*2yz
$$
And each variable is a function of variable t:

$x = t-1$; $y = t^2$; $z = \frac{1}{t}$

Let's calculate derivatives with respect to $t$:

$\frac{dx}{dt}=1$; $\frac{dy}{dt}=2t$; $\frac{dz}{dt}=-t^2$

In this situation if we want to calculate full derivative of f(x,y,z) with respect to $t$ using chain rule:
$$
\frac{df}{dt} = \frac{\delta{f}}{\delta{x}}*\frac{dx}{dt}+\frac{\delta{f}}{\delta{y}}*\frac{dy}{dt}+\frac{\delta{f}}{\delta{z}}*\frac{dz}{dt}
$$
$$
\frac{df}{dt} = cos(t-1)*e*(1)+\frac{sin(t-1)*e*(2t)}{t^2}-\frac{sin(t-1)*e*2t}{t^2} = e*cos(t-1)
$$

## Jacobian
Jacobian is a vector of partial derivatives with respect to every variable of function. This vector which size is equal to number of variables gives us information about direction of changing value of function in each point of function space. Actually it gives us a method which called as steepest decent.

Let's consider function from previous example:
$$
f(x,y,z) = sin(x)e^{yz^2}
$$
Jacobian vector is:
$$
J = [cos(x)*e^{yz^2},sin(x)*e^{yz^2}*z^2,sin(x)*e^{yz^2}*2yz]
$$

Let's consider other function for 2D space where we can map our Jacobian vector to original function:
$$
z(x,y) = 3(1-x)^2e^{-x^2-(y+1)^2}-10(\frac{x}{5}-x^3-y^5)e^{-x^2-y^2}-\frac{1}{3}e^{-(x+1)^2-y^2}
$$

## Hessian
Hessian is a matrix and it is an extention of Jacobian. If Jacobian is a vector where elements are first derivatives with respect to every variable, Hessian is a matrix where source of lines is Jacobian vector and each line is derivative of Jacobian vector with respect to one of the variables. Hence Hessian has N-to-N dimension where N - is a number of variables.

Let's consider previous example and calculate Hessian matrix:
$$
H =
\begin{bmatrix}
-sin(x)e^{yz^2} & cos(x)e^{yz^2}z^2 & cos(x)e^{yz^2}2yz \\
cos(x)e^{yz^2}z^2 & sin(x)e^{yz^2}z^4 & sin(x)e^{yz^2}2yz^3 \\
cos(x)e^{yz^2}2yz & sin(x)e^{yz^2}2yz^3 & sin(x)e^{yz^2}4y^2z^2
\end{bmatrix}
$$
    Note: this matrix is symmetrical, but this feature is valid for continuous function

Some remarks about Hessian:
* If determine of Hessian is positive therefore we have a minimum or maximum
* If first element (left right angle) of Hessian is positive therefore we are around a minimum
* If determine of Hessian is negative therefore it is not a min/max. Gradient is flat
